# Histogram

# Scatter plot

```python
# used to compare the variables
# how much one is affecting other to build a relation
x = [2,2.5,3,3.5,4,4.5,5]
y = [8,8.5,9,9.5,10,10.1,10.2]

x1 = [8,8.2,8.3,8.6,9]
y1 = [3,3.1,3.4,3.8,4.2]

plt.scatter(x,y, color = 'r',label = 'plot')
plt.scatter(x1,y1, color = 'g',label = 'plot2')
plt.xlabel("income")
plt.ylabel("savings")
plt.title("scatter plot")
plt.legend()
plt.show()
```

```python
import pandas as pd
iris = pd.read_csv('Iris.csv')
iris.head()
```

```python
for species,irissubset in iris.groupby('Species'):
  plt.scatter(irissubset['PetalLengthCm'],irissubset['PetalWidthCm'], label = species)
plt.legend()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for species,irissubset in iris.groupby('Species'):
  ax.scatter(irissubset['PetalLengthCm'],irissubset['PetalWidthCm'],irissubset['SepalWidthCm'], label = species)
plt.legend()
ax.set_xlabel("PetalLength")
ax.set_ylabel("PetalWidth")
ax.set_zlabel("SepalWidth")

plt.show()
```
