```python
class Employee:
  def __init__(self, name, id):
    self.id = id
    self.name = name

  def display(self):
    print("Name %s \n ID: %d" %(self.name, self.id))

emp = Employee("Harry", 1001)
emp1 = Employee("Smith", 1002)

emp.display()
emp1.display()
```
