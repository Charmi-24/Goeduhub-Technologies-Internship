```python
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt
```

```python
df=pd.DataFrame({ 'x1':[12,20,28,18,29,33,24,45,43,51,52,55,53,54,61,64,69,52,6,62,45,60,59,8], 
                 'x2':[39,36,30,52,54,46,55,59,63,70,66,63,58,14,8,19,7,24,77,65,80,4,72,45]})
```

```python
from sklearn.cluster import KMeans
X = df
X.head()
```

```python
km = KMeans(n_clusters=3)  # 3 clusters
```

```python
km.fit(X)
```

```python
km.cluster_centers_
```

```python
km.labels_
```

```python
df['kmeans predicted labels'] = km.labels_
df.head()
```

```python
labels = km.labels_
labels
```

```python
plt.scatter(df['x1'], df['x2'],30, labels)
plt.show()

# x1,x2,x3,x4....xn take any two features in case of multiple features
```

```python
plt.scatter(df['x1'], df['x2'],30, labels)
for i in km.cluster_centers_:
  plt.scatter(i[0],i[1], 100)

plt.show()
```

```python
# Elbow Method - for selecting the appropriate number of clusters

wss = []
cluster_range = range(1,10)

for k in cluster_range:
  km = KMeans(n_clusters=k)
  km.fit(X)
  inertia = km.inertia_
  wss.append(inertia)

print(wss)
```

```python
plt.xlabel('k - number of clusters')
plt.ylabel('Sum of squared error')

plt.plot(cluster_range,wss)
plt.show()
```
