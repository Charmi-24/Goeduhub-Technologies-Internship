# range function

```python
# range(start, stop, step) # default start from 0, default step 1
```

```python
for i in range(5): # only one index means -- stop 
  print(i)
```

```python
n=2
for i in range(1,11):   #start, stop
  t = n*i
  print(n,"*", i,"=",t)  
```

```python
n=2
for i in range(1,11,2):   #start, stop, step
  t = n*i
  print(n,"*", i,"=",t)
```

# star pattern

```python
r = int(input("enter the no of rows"))
for i in range(0,r):
  for j in range(i):
    print("*",end="")
  print()
```

```python
for i in range(6,0,-1):
  print(i)
```

```python
for i in range(5):
  if(i==3):
    continue
  print(i)
```
