***Task - Predicting a Startups Profit/Success Rate using Multiple Linear Regression in Python***

Here 50 startups dataset containing 5 columns  like “R&D Spend”, “Administration”, “Marketing Spend”, “State”, “Profit”.


In this dataset first 3 columns provides you spending on Research , Administration and Marketing respectively. State indicates startup based on that state. Profit indicates how much profits earned by a startup.


Clearly, we can understand that it is a multiple linear regression problem, as the independent variables are more than one.


Prepare a prediction model for profit of 50_Startups data in Python

```python
# Importing Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn import metrics
```

```python
# Loading the data
df = pd.read_csv("/content/50_Startups.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
# Dealing the null values
df = df.replace(0, np.nan)
df['R&D Spend'].fillna(df['R&D Spend'].median(),inplace=True)
df['Marketing Spend'].fillna(df['Marketing Spend'].median(),inplace=True)
```

```python
df['State'].unique()
```

```python
'''
Dummy Variables / One Hot Encoding
var = pd.get_dummies(df['State'])
df = df.join(var) OR merge = pd.concat([df,var], axis = 'columns')
df.head()
df.drop('State', axis = 1, inplace = True)
df.drop(['New York'], axis = 'columns', inplace = True)


Label Encoding
from sklearn.processing import LabelEncoder
le = LabelEncoder()
newdf = df
s = le.fit_transform(newdf['State'])
newdf.head()

'''
```

```python
# Target Encoding for categorical variable

targ_en = df.groupby('State')['Profit'].mean()
df = df.merge(targ_en, how = 'left', on = 'State')
df.drop('State', axis = 1, inplace = True)
df = df.rename(columns = {'Profit_x': 'Profit', 'Profit_y': 'State'})
```

```python
df.head()
```

```python
# Assigning the variables
X = df.drop('Profit', axis = 1)
y = df['Profit']
```

```python
# Creating a model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.15, random_state = 5)
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train,y_train)
```

```python
# Predictions
y_pred = lm.predict(X_test)
```

```python
sklearn.metrics.r2_score(y_test, y_pred)
```

```python
metrics.mean_squared_error(y_test,y_pred)
```

```python
# Error
np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```
