1. Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a carâ€™s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.
         Days=[1,2,3,4,5,6,7,8]

         Speed=[60,62,61,58,56,57,46,63]

```python
from matplotlib import pyplot as plt
Days=[1,2,3,4,5,6,7,8]
Speed = [60,62,61,58,56,57,46,63]
plt.plot(Days,Speed)
plt.title("Car Speed Measurement")
plt.xlabel("Days")
plt.ylabel("Car speed")
plt.show()
```

```python
#plt.plot(Days,Speed,color ='green', linestyle='dotted', marker='o', markerfacecolor = 'blue', markersize=12)
plt.plot(Days,Speed,'g:o', mfc = 'blue', ms = 12)
plt.title("Car Speed Measurement")
plt.xlabel("Days")
plt.ylabel("Car speed")
plt.show()
```

4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.

```python
import numpy as np
base = np.arange(0, 6, 0.2);
sf = np.sin(base)
cf = np.cos(base)

plt.plot(base,sf, 'b--', label = 'sin')
plt.plot(base, cf, 'm--', label = 'cosine')
plt.legend()
plt.show()
```

5. Plot Simple bar chart showing popularity of Programming Languages.

Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']

Popularity = [56, 39, 34, 34, 29]

Security = [44 ,36 ,55, 50, 42]

Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.

```python
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]

x = np.arange(len(Languages))
width = 0.3
plt.bar(x - width/2, Popularity, width, label="Popularity",color='y')
plt.bar(x + width/2,Security,width, label="Security",color='b')

plt.legend()
plt.xlabel('Languages')
plt.ylabel('Scale')
plt.xticks(x, Languages)
plt.title('Multiple Bar Chart')
plt.show()
```

```python
x = np.arange(len(Languages))
width = 0.3
plt.barh(x - width/2, Popularity, width, label="Popularity",color='y')
plt.barh(x + width/2,Security,width, label="Security",color='b')

plt.legend()
plt.xlabel('Scale')
plt.ylabel('Languages')
plt.yticks(x, Languages)
plt.title('Multiple Bar Chart')
plt.show()
```
